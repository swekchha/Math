```python
#Question no 2
import numpy as np
import math
import random
import matplotlib.pyplot as plt

random.seed(7)

given_mean = 10
given_std = 0.5
sum_val = 0

random_num = np.random.normal(given_mean, given_std, 500)
a = list(random_num)


for i in random_num:
    sum_val = sum_val + i

mean = sum_val / len(a)
a.sort()

if len(a) % 2 == 0:
    even_start = a[len(a) // 2 - 1]
    even_end = a[len(a) // 2]
    median = (even_start + even_end) / 2
else:
    median = a[len(a) // 2]

half_equation = sum((x - mean)**2 for x in a)
std = math.sqrt(half_equation / (len(a) - 1))

print('Mean is:', mean)
print('Median is:', median)
print('Standard Deviation:', std)

plt.hist(random_num, bins=10, color='pink', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of 500 Random Numbers')
plt.grid(True)
```
